```python
def drawHistogram(attribute, title, labels, xLabel, yLabel, partition):
  
  fig, (ax) = plt.subplots(nrows=1, ncols=1, figsize=(9, 4))

  ax.yaxis.grid(True)
  ax.set_ylabel(yLabel)
  ax.set_xlabel(xLabel)
  ax.set_title(title)
  ax.hist(attribute, partition, label=labels, histtype='bar')
  ax.legend()

  plt.legend()
  plt.show()


drawHistogram([data.situacao_basico_ano1, data.situacao_basico_ano2, data.situacao_basico_ano3, data.situacao_basico_ano4],
              'Aprovação dos alunos -  Ensino Básico',
              ['Ano 1', 'Ano 2', 'Ano 3', 'Ano 4'],
              'Anos',
              'Número de alunos',
              3)

drawHistogram([data.situacao_tecnico_ano1, data.situacao_tecnico_ano2, data.situacao_tecnico_ano3, data.situacao_tecnico_ano4],
              'Aprovação dos alunos -  Ensino Técnico',
              ['Ano 1', 'Ano 2', 'Ano 3', 'Ano 4'],
              'Anos',
              'Número de alunos',
              3)

drawHistogram([data.permaneceu_ano2, data.permaneceu_ano3, data.permaneceu_ano4],
              'Permanência dos Alunos no Curso',
              ['Ano 2', 'Ano 3', 'Ano 4'],
              'Anos',
              'Número de alunos',
              3)

drawHistogram(data.media_basico_ano1[~np.isnan(data.media_basico_ano1)],
              'Histograma Média das Notas - Ensino Básico Ano 1',
              ['Número de alunos'],
              'Médias dos alunos',
              'Número de alunos',
              50)

drawHistogram(data.media_tecnico_ano1[~np.isnan(data.media_tecnico_ano1)],
              'Histograma Média das Notas - Ensino Técnico Ano 1',
              ['Número de alunos'],
              'Médias dos alunos',
              'Número de alunos',
              50)

drawHistogram(data.media_basico_ano2[~np.isnan(data.media_basico_ano2)],
              'Histograma Média das Notas - Ensino Básico Ano 2',
              ['Número de alunos'],
              'Médias dos alunos',
              'Número de alunos',
              50)

drawHistogram(data.media_tecnico_ano2[~np.isnan(data.media_tecnico_ano2)],
              'Histograma Média das Notas - Ensino Técnico Ano 2',
              ['Número de alunos'],
              'Médias dos alunos',
              'Número de alunos',
              50)

drawHistogram(data.media_basico_ano3[~np.isnan(data.media_basico_ano3)],
              'Histograma Média das Notas - Ensino Básico Ano 3',
              ['Número de alunos'],
              'Médias dos alunos',
              'Número de alunos',
              50)

drawHistogram(data.media_tecnico_ano3[~np.isnan(data.media_tecnico_ano3)],
              'Histograma Média das Notas - Ensino Técnico Ano 3',
              ['Número de alunos'],
              'Médias dos alunos',
              'Número de alunos',
              50)

drawHistogram(data.media_basico_ano4[~np.isnan(data.media_basico_ano4)],
              'Histograma Média das Notas - Ensino Básico Ano 4',
              ['Número de alunos'],
              'Médias dos alunos',
              'Número de alunos',
              50)

drawHistogram(data.media_tecnico_ano4[~np.isnan(data.media_tecnico_ano4)],
              'Histograma Média das Notas - Ensino Técnico Ano 4',
              ['Número de alunos'],
              'Médias dos alunos',
              'Número de alunos',
              50)

```

```python
def drawBoxplot(boxplotData, title, yLabel):
  fig, (ax) = plt.subplots(nrows=1, ncols=1, figsize=(9, 4))

  ax.set_title(title)
  ax.yaxis.grid(True)
  ax.set_ylabel(yLabel)

  labels = ["Ano 1", "Ano 2", "Ano 3", "Ano 4"]
  boxplot = ax.boxplot(boxplotData, patch_artist=True, labels=labels)

  # fill with colors
  colors = ['pink', 'lightblue', 'lightgreen', 'yellow']
  for patch, color in zip(boxplot['boxes'], colors):
    patch.set_facecolor(color)

  plt.show()

boxplotData = [data.media_basico_ano1[~np.isnan(data.media_basico_ano1)], 
               data.media_basico_ano2[~np.isnan(data.media_basico_ano2)],
               data.media_basico_ano3[~np.isnan(data.media_basico_ano3)],
               data.media_basico_ano4[~np.isnan(data.media_basico_ano4)]]
drawBoxplot(boxplotData, 'Media das disciplinas - Ensino Básico', 'Média das Notas [0..10]')

boxplotData = [data.media_tecnico_ano1[~np.isnan(data.media_tecnico_ano1)], 
               data.media_tecnico_ano2[~np.isnan(data.media_tecnico_ano2)],
               data.media_tecnico_ano3[~np.isnan(data.media_tecnico_ano3)],
               data.media_tecnico_ano4[~np.isnan(data.media_tecnico_ano4)]]
drawBoxplot(boxplotData, 'Media das disciplinas - Ensino Técnico', 'Média das Notas [0..10]')

boxplotData = [data.faltas_basico_ano1[~np.isnan(data.faltas_basico_ano1)], 
               data.faltas_basico_ano2[~np.isnan(data.faltas_basico_ano2)],
               data.faltas_basico_ano3[~np.isnan(data.faltas_basico_ano3)],
               data.faltas_basico_ano4[~np.isnan(data.faltas_basico_ano4)]]
drawBoxplot(boxplotData, 'Media das Faltas - Ensino Básico', 'Média das Faltas')

boxplotData = [data.faltas_tecnico_ano1[~np.isnan(data.faltas_tecnico_ano1)], 
               data.faltas_tecnico_ano2[~np.isnan(data.faltas_tecnico_ano2)],
               data.faltas_tecnico_ano3[~np.isnan(data.faltas_tecnico_ano3)],
               data.faltas_tecnico_ano4[~np.isnan(data.faltas_tecnico_ano4)]]
drawBoxplot(boxplotData, 'Media das Faltas - Ensino Técnico', 'Média das Faltas')
```

Medidas estatísticas

```python
def printStatistics(atributte, atributteName):
  print("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}".format(  
        atributteName,
        atributte.mean(), 
        atributte.median(),  
        atributte.quantile(q=0.75), 
        #atributte.mode(),
        atributte.var(), 
        atributte.std(), 
        atributte.skew(),
        atributte.kurtosis()))
  
def printMode(atributte, atributteName):
  print("{}\n{}\n\n".format(atributteName, atributte.mode()))

print("atributo\tmédia\tmediana\t3º quartil\tvariância\tdesvio padrão\tobliquidade\tcurtose")
printStatistics(data.media_basico_ano1[~np.isnan(data.media_basico_ano1)], "media_basico_ano1")
printStatistics(data.media_tecnico_ano1[~np.isnan(data.media_tecnico_ano1)], "media_tecnico_ano1")
printStatistics(data.faltas_basico_ano1[~np.isnan(data.faltas_basico_ano1)], "faltas_basico_ano1")
printStatistics(data.faltas_tecnico_ano1[~np.isnan(data.faltas_tecnico_ano1)], "faltas_tecnico_ano1")
printStatistics(data.media_basico_ano2[~np.isnan(data.media_basico_ano2)], "media_basico_ano2")
printStatistics(data.media_tecnico_ano2[~np.isnan(data.media_tecnico_ano2)], "media_tecnico_ano2")
printStatistics(data.faltas_basico_ano2[~np.isnan(data.faltas_basico_ano2)], "faltas_basico_ano2")
printStatistics(data.faltas_tecnico_ano2[~np.isnan(data.faltas_tecnico_ano2)], "faltas_tecnico_ano2")
printStatistics(data.media_basico_ano3[~np.isnan(data.media_basico_ano3)], "media_basico_ano3")
printStatistics(data.media_tecnico_ano3[~np.isnan(data.media_tecnico_ano3)], "media_tecnico_ano3")
printStatistics(data.faltas_basico_ano3[~np.isnan(data.faltas_basico_ano3)], "faltas_basico_ano3")
printStatistics(data.faltas_tecnico_ano3[~np.isnan(data.faltas_tecnico_ano3)], "faltas_tecnico_ano3")
printStatistics(data.media_basico_ano4[~np.isnan(data.media_basico_ano4)], "media_basico_ano4")
printStatistics(data.media_tecnico_ano4[~np.isnan(data.media_tecnico_ano4)], "media_tecnico_ano4")
printStatistics(data.faltas_basico_ano4[~np.isnan(data.faltas_basico_ano4)], "faltas_basico_ano4")
printStatistics(data.faltas_tecnico_ano4[~np.isnan(data.faltas_tecnico_ano4)], "faltas_tecnico_ano4")

print("\n\nModa")
printMode(data.media_basico_ano1[~np.isnan(data.media_basico_ano1)], "media_basico_ano1")
printMode(data.media_tecnico_ano1[~np.isnan(data.media_tecnico_ano1)], "media_tecnico_ano1")
printMode(data.faltas_basico_ano1[~np.isnan(data.faltas_basico_ano1)], "faltas_basico_ano1")
printMode(data.faltas_tecnico_ano1[~np.isnan(data.faltas_tecnico_ano1)], "faltas_tecnico_ano1")
printMode(data.media_basico_ano2[~np.isnan(data.media_basico_ano2)], "media_basico_ano2")
printMode(data.media_tecnico_ano2[~np.isnan(data.media_tecnico_ano2)], "media_tecnico_ano2")
printMode(data.faltas_basico_ano2[~np.isnan(data.faltas_basico_ano2)], "faltas_basico_ano2")
printMode(data.faltas_tecnico_ano2[~np.isnan(data.faltas_tecnico_ano2)], "faltas_tecnico_ano2")
printMode(data.media_basico_ano3[~np.isnan(data.media_basico_ano3)], "media_basico_ano3")
printMode(data.media_tecnico_ano3[~np.isnan(data.media_tecnico_ano3)], "media_tecnico_ano3")
printMode(data.faltas_basico_ano3[~np.isnan(data.faltas_basico_ano3)], "faltas_basico_ano3")
printMode(data.faltas_tecnico_ano3[~np.isnan(data.faltas_tecnico_ano3)], "faltas_tecnico_ano3")
printMode(data.media_basico_ano4[~np.isnan(data.media_basico_ano4)], "media_basico_ano4")
printMode(data.media_tecnico_ano4[~np.isnan(data.media_tecnico_ano4)], "media_tecnico_ano4")
printMode(data.faltas_basico_ano4[~np.isnan(data.faltas_basico_ano4)], "faltas_basico_ano4")
printMode(data.faltas_tecnico_ano4[~np.isnan(data.faltas_tecnico_ano4)], "faltas_tecnico_ano4")
```
